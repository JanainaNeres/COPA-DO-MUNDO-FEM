# MAPAS

Definições básicas para as construções dos mapas

```python
from vega_datasets import data
import altair as alt
 
terra = alt.topo_feature(data.world_110m.url,'countries') #DataBase
sphere = alt.sphere()
graticule = alt.graticule()
mar = alt.Chart(sphere).mark_geoshape(fill='#B9EDE6')
#linhas = alt.Chart(graticule).mark_geoshape(stroke='#3DD1F2', strokeWidth=0.5)
fundo = alt.Chart(terra).mark_geoshape(fill='#E6E6FA',stroke="black", strokeWidth=0.1,)

mar+fundo 
```

# MAPA COPA DO MUNDO FEM  1991

```python
# 1991
data= {
  "Alemanha" :  [4, 'Semi-Final' ],
  "Brasil" :  [9,'Oitavas de Final'],
  "China" :[5, 'Oitavas de Final'],
  "Dinamarca" : [7, 'Quartas de Final'],
  "Estados Unidos" :  [1, 'Campeã'],
  "Itália" : [6, 'Quartas de Final'],
  "Japão" : [12, 'Oitavas de Final'],
  "Nigéria" : [10, 'Oitavas de Final'],
  "Noruega" : [2, 'Vice Campeã'],
  "Nova Zelândia" : [11, 'Oitavas de Final'],
  "Suécia" : [3, 'Semi-Final'],
  "Taiwan" : [8,  'Quartas de Final']
    
}

df = pd.DataFrame(data, index = ['1991', 'E1'])
data1 = {'Selecao' : list(df.columns),
         'Posicao': list(df.loc['1991']),
         'E1': list(df.loc['E1'])       
          }
df0 = pd.DataFrame(data1)


alpha2=[]
for a in df0.Selecao:
    alpha2.append(mapa1[a])
df0['alpha2'] = alpha2
df1= df0.merge(ISO,on='alpha2')


# Mapa 
mapa = alt.Chart(terra
         ).mark_geoshape(stroke="#666666", strokeWidth=0.1).transform_lookup(
    lookup = 'id',from_=alt.LookupData(data=df1, key='id', fields=['Posicao','Selecao','E1'])
    ).encode(
    color = alt.Color('E1:N', 
                      scale=alt.Scale(domain=df1.E1.unique(),range=[ '#ff48a5', #semi-finaL
                                                                     '#ffaed7', #oitavas
                                                                     '#ff77bc', #quartas 
                                                                     '#A834A6',#1
                                                                     '#E36BCD', #2 
                                                                     '#ffcae5' #fase de grups
                                                                  ]
                                     ),#scheme=cor),
                      sort='ascending',
                      legend=alt.Legend(title='Colocações',
                                        direction='horizontal',
                                        orient='bottom-right',
                                        fillColor='lightgrey',
                                        padding=9,
                                        symbolStrokeWidth=1,
                                        columns=3,
                                        offset=2,
                                        labelFontSize=10,
                                        titleFontSize=15,
                                        labelFont='Courier',
                                        titleFont='Courier',
                                        titleLimit=300,
                                        strokeColor='black',
                                        values=['Campeã','Vice Campeã','Semi-Final','Quartas de Final','Oitavas de Final']
                      )
                      ),
    #range=['gold','red']
    tooltip = [alt.Tooltip('Selecao:N',title = 'Country'),alt.Tooltip('Posicao:Q'), alt.Tooltip('E1:N')]
    )

fig = (fundo+mapa).properties(width=800, height=400,title={'text':['Copa do Mundo Feminina,', 'China, 1991.',],
                                                               'subtitle': ['Classificação de cada seleção ',
                                                                            'na Copa do Mundo']},
                        ).project('equirectangular',#center= [15,5],scale=130,clipExtent= [[0, 0], [800, 400]]
                                 ).configure_title(fontSize=15,
                                                                  font='Courier',
                                                                  subtitleFont='Courier',
                                                                  subtitleFontSize=12,
                                                                  anchor='start',
                                                                  color='black',
                                                                  orient='bottom',
                                                                  angle=80,
                                                                  dx=5,dy=-100
                                                                 ).configure_view(strokeWidth=0)

fig

```
